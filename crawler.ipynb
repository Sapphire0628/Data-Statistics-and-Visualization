```python
import bs4
import csv
import requests
from lxml import etree
from datetime import datetime
```

# Medium News

```python
url = 'https://medium.com/towards-data-science/newsletter'
response = requests.get(url)

soup = bs4.BeautifulSoup(response.content, 'html.parser')
tree = etree.HTML(response.content)

news_data = []

for index, news_list in enumerate(soup.find_all('div', {'class': 'if ab gm'})):
    news = news_list.find("h2").text
    date_str = news_list.find("h3").text
    
    # Convert date string to datetime object
    try:
        date = datetime.strptime(date_str, '%b %d, %Y')
    except ValueError:
        # If parsing fails, use the original string
        date = date_str
    
    news_data.append([index+1, news, date])
    print(f"News Articles {index+1}: {news} , Date : {date}")

# Save data to CSV file
csv_filename = './dataset/Medium_news.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Index', 'News', 'Date'])  # Write header
    csv_writer.writerows(news_data)

print(f"Data saved to {csv_filename}")
```

# MIT News

```python
news_data = []
index = 1
for page in range(1,121): 
    url = f'https://news.mit.edu/topic/computers?page={page}'
    response = requests.get(url)

    soup = bs4.BeautifulSoup(response.content, 'html.parser')
    tree = etree.HTML(response.content)

    for i, news_list in enumerate(soup.find("div", {'class' : 'page-term--views--list'}).find_all('div', {'class':'term-page--news-article--item--descr'})):
        news = news_list.find('span').text
        date_str = news_list.find('time').text
        
        # Convert date string to datetime object
        try:
            date = datetime.strptime(date_str, '%B %d, %Y')
        except ValueError:
            # If parsing fails, use the original string
            date = date_str
        
        news_data.append([index, news, date])
        print(f"News Articles {index}: {news} , Date : {date}")
        index +=1

# Save data to CSV file
csv_filename = './dataset/MIT_news.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Index', 'News', 'Date'])  # Write header
    csv_writer.writerows(news_data)

print(f"Data saved to {csv_filename}")

```

# ScienceNews

```python
news_data = []
index = 1
for page in range(1,18): 
    url = f'https://www.sciencenews.org/topic/computing/page/{page}'
    response = requests.get(url)


    soup = bs4.BeautifulSoup(response.content, 'html.parser')
    tree = etree.HTML(response.content)

    for i, news_list in enumerate(soup.find('ol', {'class':'river-with-sidebar__list___5dLsp'}).find_all('li')):
        news = news_list.find('h3').text.replace('\n','').replace('\t','')
        date_str = news_list.find('time').text
        
        # Convert date string to datetime object
        try:
            date = datetime.strptime(date_str, '%B %d, %Y')
        except ValueError:
            # If parsing fails, use the original string
            date = date_str
        
        news_data.append([index, news, date])
        print(f"News Articles {index}: {news} , Date : {date}")
        index +=1


# Save data to CSV file
csv_filename = './dataset/ScienceNews.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Index', 'News', 'Date'])  # Write header
    csv_writer.writerows(news_data)

print(f"Data saved to {csv_filename}")
```

# Nature

```python
import time

index = 1

for page in range(1,21): 

    url = f'https://www.nature.com/search?q=computer-science&order=date_asc&date_range=2024-2025&page={page}'
    response = requests.get(url)


    soup = bs4.BeautifulSoup(response.content, 'html.parser')
    tree = etree.HTML(response.content)

    for i, news_list in enumerate(soup.find('ul', {'class':'app-article-list-row'}).find_all('li', {'class':'app-article-list-row__item'})):
        try:
            news = news_list.find('h3').text.replace('\n','')
            date_str = news_list.find('time').text
        
        # Convert date string to datetime object
        
            date = datetime.strptime(date_str, '%d %b %Y')
        except ValueError:
            # If parsing fails, use the original string
            continue
        
        news_data.append([index, news, date])
        print(f"News Articles {index}: {news} , Date : {date}")
        index +=1
    print("Finish Page ",page )
    time.sleep(1)

# Save data to CSV file
csv_filename = './dataset/Nature_research.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Index', 'News', 'Date'])  # Write header
    csv_writer.writerows(news_data)

print(f"Data saved to {csv_filename}")
        
```
